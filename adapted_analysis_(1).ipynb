# Section 2 of adpted data

```python
df = pd.read_csv("/content/Maria - adapted_section2_green.csv")
df.shape
```

```python
df.isnull().sum()
```

# Deleting the columns that have too much null values

```python
df.drop(['Policy milestone 1','Policy Milestone 2','Policy milestone 4','Policy milestone 3'],axis=1,inplace=True)
```

```python
df.shape
# after deleting null columns ,four columns remain for analysis
```

```python
print('unique values in policy file : ',df['policy file '].nunique())
# in policy file, every values is different from each other
print('unique values in Policy stage: ',df['Policy stage'].nunique())
# in required EO based information 1 , almost every values is different from each other
print('unique values in required EO based information 1 : ',df['required EO based information 1'].nunique())
print('unique values in USER 1 : ',df['USER 1'].nunique())
```

```python
df.head(1)
```

```python
# @title Policy stage vs USER 1

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['USER 1'].value_counts()
    for x_label, grp in df.groupby('Policy stage')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Policy stage')
_ = plt.ylabel('USER 1')
```

```python
# @title USER 1

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('USER 1').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

# **Policy stage**

```python
# @title Policy stage

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Policy stage').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
